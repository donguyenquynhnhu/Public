```python
import pandas as pd
import numpy as np

data = {
    "ID" : [101,102,103,104,105],
    "Name" : ["Quynh Nhu", "Nhat Hao", None, "Minh Quan", "Huu Duc"],
    "Age" : [26, 18, 20, None, 19],
    "Salary" : [5000, 35000, 65000, 10000, None]
}

df = pd.DataFrame(data)
df

#Dien gia tri thieu trong cot Age
min_age = df['Age'].min()
df["Age"] = df['Age'].fillna(min_age)

#Loại bỏ các dòng chứa dữ liệu thiếu
df.dropna(inplace=True)

#Dien gia tri thieu trong Name = "unknown"ArithmeticError
df["Name"] = df["Name"].fillna("unknown")

#Dien gia tri thieu trong salary = phuong phap Interpolation 
df["Salary"] = df["Salary"].interpolate()

print(df)
```

```python
#cho data
import pandas as pd

# Creating the first DataFrame
df1 = pd.DataFrame({
    "ID": [101, 102, 103],
    "Score_A": [70, 80, 90]
})

# Creating the second DataFrame
df2 = pd.DataFrame({
    "ID": [102, 103, 104],
    "Score_B": [85, 95, 75]
})

# Merging the DataFrames using an inner join
merged_df = pd.merge(df1, df2, on="ID", how="inner")

# Display the merged DataFrame
print(merged_df)
```

```python
import pandas as pd
import numpy as np

# Bước 1: Tạo DataFrame
data = pd.DataFrame({
    "ID": range(1, 100001),  # ID từ 1 đến 100,000
    "Value": np.random.randint(1, 1000000, 100000)  # Số nguyên ngẫu nhiên từ 1 đến 1,000,000
})

# Hiển thị vài dòng đầu tiên của DataFrame
print("DataFrame:")
print(data.head())

# Bước 2: Sử dụng .astype() để chuyển đổi kiểu dữ liệu
data['Value'] = data['Value'].astype(np.int32)  # Chuyển đổi 'Value' thành int32

# Kiểm tra kiểu dữ liệu
print("\nKiểu Dữ Liệu:")
print(data.dtypes)

# Bước 3: Tìm giá trị duy nhất trong cột 'Value'
unique_values = data['Value'].unique()

# Hiển thị giá trị duy nhất
print("\nGiá Trị Duy Nhất trong Cột 'Value' (Mẫu):")
print(unique_values[:10])  # Chỉ hiển thị 10 giá trị đầu tiên cho ngắn gọn

# Bước 4: Lọc các hàng mà 'Value' lớn hơn 90
filtered_data = data.query('Value > 90')
print(filtered_data.head())  # Hiển thị vài dòng đầu tiên của DataFrame đã lọc
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 100)
y1 = np.sin(t)
y2 = np.cos(t)
y3 = np.sin(t) * np.cos(t)

#Tạo biểu đồ các đường theo thời gian
plt.figure(figsize=(10,6)) #đặt kích thước bảng
plt.plot(t, y1, label = 'sin(t)', color = "blue")
plt.plot(t, y2, label = 'cos(t)', color = "red")
plt.plot(t, y3, label = 'sint(t) * cos(t)', color = "orange")

#Thêm thành phần cho biểu đồ
plt.title("Biểu đồ Hàm số theo thời gian")
plt.xlabel("Thời gian (t)")
plt.ylabel('Giá Trị')                                
plt.legend()                                         
plt.grid()                                         
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dữ liệu
labels = ["Q1", "Q2", "Q3", "Q4"]
data = [500, 350, 600, 620]

# Vẽ biểu đồ nhóm
x = np.arange(len(labels))  # Vị trí của các thanh
width = 0.35  # Độ rộng của các thanh

fig, ax = plt.subplots()
bars = ax.bar(x, data, width, label='Doanh thu')

# Thêm nhãn và tiêu đề
ax.set_ylabel('Doanh thu')
ax.set_title('Doanh thu của A, B, C trong 4 quý')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()

#Biểu đồ tròn
# Dữ liệu biểu đồ tròn
labels = ["A", "B", "C", "D"]
sizes = [25, 25, 25, 25]  # Tỷ lệ phần trăm cho mỗi phần
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffff99"]  # Màu sắc cho các phần

# Vẽ biểu đồ tròn
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Đảm bảo hình tròn
plt.title('Biểu đồ tròn hiển thị tỷ lệ phần trăm của 4 công ty')
plt.show()

#Biểu đồ phân tán
# Dữ liệu biểu đồ phân tán
x = np.random.rand(100) * 100  # 100 điểm ngẫu nhiên trên trục x
y = np.random.rand(100) * 100  # 100 điểm ngẫu nhiên trên trục y
colors = np.random.rand(100) * 300  # Màu sắc ngẫu nhiên cho các điểm

plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=colors, alpha=0.5) 
plt.title('Biểu đồ phân tán')
plt.xlabel('Trục X')
plt.ylabel('Trục Y')
plt.colorbar()  
plt.show()

#Biểu đồ nhiệt
data = np.random.rand(10, 10)
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Biểu Đồ Nhiệt Thể Hiện Sự Phân Bố Dữ Liệu')
plt.xlabel('Trục X')
plt.ylabel('Trục Y')
plt.show()
```
