```python
import pandas as pd

# Tạo dataFrame với dictionary đúng cấu trúc
data = {
    "name": ["Hao","Nhu","Quan","Duc"],
    "age": [21,22,23,24],
    "city": ["HCMC","Soctrang","CaMau","Hue"]
}

# Tạo DataFrame
df = pd.DataFrame(data)

# In DataFrame
print(df)
# In cot City
print(df["city"])
# in tuoi tren 22
print(df[df["age"] >= 22])
# Tinh tong tuoi tb
print(df["age"].mean())
```

```python
import matplotlib.pyplot as plt
#Bieu do duong
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x,y,label = "bieu do duong", color = "darkblue", marker = "o")
plt.title("Bieu do duong")
plt.xlabel("Truc x")
plt.ylabel("Truc y")
plt.legend()
plt.show()

#bieu do cot
ten = ["Hao","Quan","Nhu","Duc"]
diem = [1,2,3,4]
plt.bar(ten, diem, label = "bieu do cot", color = "orange")
plt.title("bieu do cot")
plt.xlabel("Truc x")
plt.ylabel("Truc y")
plt.show()
```
